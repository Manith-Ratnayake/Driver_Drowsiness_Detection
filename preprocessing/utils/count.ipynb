```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict


def plot_combined_csvs(root_dir, save_path=None, show_plot=True):
    combined_counts = defaultdict(int)
    csv_files = []

    # find all csv files
    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            if file.endswith(".csv"):
                csv_files.append(os.path.join(dirpath, file))

    if not csv_files:
        print("No CSV files found.")
        return

    csv_files.sort()

    # read and merge all csvs
    for csv_path in csv_files:
        print(f"Reading: {csv_path}")

        df = pd.read_csv(csv_path)

        if "square_size" not in df.columns or "count" not in df.columns:
            print(f"Skipping {csv_path} (invalid format)")
            continue

        for _, row in df.iterrows():
            size = int(row["square_size"])
            count = int(row["count"])
            combined_counts[size] += count

    # convert to dataframe
    combined_df = pd.DataFrame({
        "square_size": list(combined_counts.keys()),
        "count": list(combined_counts.values())
    })

    combined_df = combined_df.sort_values("square_size")

    # plot
    plt.figure(figsize=(14, 6))
    plt.bar(combined_df["square_size"], combined_df["count"], width=0.9)

    plt.title("Combined Eye Square Size Distribution (All CSVs)")
    plt.xlabel("Square Size")
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=200, bbox_inches="tight")
        print(f"Saved combined plot: {save_path}")

    if show_plot:
        plt.show()
    else:
        plt.close()


if __name__ == "__main__":
    root_folder = "/content/abc"  

    plot_combined_csvs(
        root_dir=root_folder,
        save_path="/kaggle/working/combined_eye_hist.png", 
        show_plot=True
    )
```
